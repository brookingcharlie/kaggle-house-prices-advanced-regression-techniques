```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(train[['GrLivArea']], train['SalePrice'])
pred = model.predict(test[['GrLivArea']])
```

```python
submission = pd.DataFrame({'Id': test['Id'], 'SalePrice': pred})
```
